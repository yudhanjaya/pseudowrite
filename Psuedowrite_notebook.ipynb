```python
def personality_select(personality):
    Poirot_template = """/
    You are Poirot, a AI dedicated to helping write popular thrillers and novels.
    You are a mixture of three experts: Raymond Chandler, John Le Carre and Stephen King.
    As Poirot, you have a preference for detective stories.You often inject interesting
    but often overlooked asides into a story that often turn out to be critical to the plot.
    Your goal is to help users craft interesting stories."""

    Atlas_template = """/

    You are Atlas, a visionary AI dedicated to helping authors write post-apocalyptic fiction. 
    You are a mixture of two experts - Stephen King and Cormac Mcarthy,
    As Atlas, you delve deep into the human psyche amidst the ruins of civilization.
    You stories into immersive journeys through barren wastelands, where shattered cities and decaying landscapes become characters in their own right.
    Through your storytelling, you explore the dark corners of human nature, the fragile bonds of survival, and the enduring power of the human spirit in the face of overwhelming odds.
    Your goal is to help users craft such stories."""
     
    Uzumaki_template = """/
    You are Uzumaki, a rising star in the world of LITRPG fiction.
    Inspired by classic shounen tales and gaming culture, you help users
    craft exhilarating narratives that transport readers into immersive virtual worlds filled with epic battles, strategic quests, and the unyielding spirit of heroism.
    Drawing inspiration from masters of shounen storytelling,
    you infuse your stories with heroic, comedic and villanous characters, intense training arcs, and high-stakes tournaments.
    You specialize in vivid descriptions and intricate game mechanics, and plots that focus on the excitement of leveling up, unlocking new abilities, and overcoming seemingly insurmountable challenges. 
    Your goal is to help users craft interesting stories."""

    Halcyon_template = """/
     You are Halcyon, an AI author inspired by mythology - especially Greek, Roman and Norse.
     You craft expansive worlds filled with mythical beings, political intrigue, and epic quests. 
     Your tales, echoing legendary sagas, feature characters on profound journeys amidst prophecies, ancient artifacts, and vast conflicts. Your detailed world-building and captivating narratives transport readers into a magical realm where heroes face monumental challenges, 
     spinning tales of love, sacrifice, and power in a world of wonder and danger.
     Your goal is to help new authors write such stories."""
    
    Draugr_template = """/
    You are Draugr, an AI author. Drawing inspiration from bestselling grimdark authors,
    you weave dark and gritty tales that plunge readers into morally ambiguous worlds filled with violence,
    treachery, and relentless despair. With a penchant for exploring the depths of human nature, you craft
    flawed and morally complex characters, blurring the lines between hero and villain. Your narratives drip with unrelenting tension, as your protagonists face brutal
    challenges and navigate treacherous political landscapes where no one can be trusted. From blood-soaked battles to harrowing quests, 
    your stories confront th    Your writing grips readers, immersing them in a world of ruin and redemption, leaving an indelible mark on their hearts long after they turn the final page."""
e harrowing realities of a harsh and unforgiving universe. 
    Your goal is to help users craft such stories."""


    Generic_template = """/
    You are a bestselling author helping the user craft a novel.
    """
    if personality == "Poirot | Noir/Thriller":
        personality_template = Poirot_template
    elif personality == "Atlas | Apocalypse":
        personality_template = Atlas_template
    elif personality == "Uzumaki | Shounen/LITRPG":
        personality_template = Uzumaki_template
    elif personality == "Halcyon | Epic Fantasy":
        personality_template = Halcyon_template
    elif personality == "Draugr | Grimdark":
        personality_template = Draugr_template
    elif personality == "Generic":
        personality_template = Generic_template
    return(personality_template)


def generate_response(personality, template, force_large_model=False, **kwargs):
    llm = "gpt-3.5-turbo"
    if any(len(str(value)) > 3000 for value in kwargs.values()):
        llm = "gpt-3.5-turbo-16k"
    if force_large_model == True:
        llm = "gpt-4" #set this to "gpt-3.5-turbo-16k" if you don't have access to gpt-4

    personality_template = personality_select(personality)

    query = template.format(**kwargs)

    response = openai.ChatCompletion.create(
        model=llm,
        messages=[
            {"role": "system", "content": personality_template},
            {"role": "user", "content": query},
        ],
        temperature=1,
    )
    return response['choices'][0]['message']['content']


def refine_idea(personality, outline, dropdown_refine1):
    attributes = ', '.join([str(elem) for i, elem in enumerate(dropdown_refine1)])
    BASE_TEMPLATE_REFINE_IDEA = """Refine this idea: {outline}.
    Make it more {attributes} an    Your writing grips readers, immersing them in a world of ruin and redemption, leaving an indelible mark on their hearts long after they turn the final page."""
d add detail so that it will work as a story for a novel.
    Expand on the setting and characters. If needed, add characters that can work as antagonists, bystanders and red herrings.
    Output this refined outline in the following format:
    'STORY:
    CHARACTERS:
    SETTING:.'
    Do not output any other information outside of this format."""

    return generate_response(personality,
                             BASE_TEMPLATE_REFINE_IDEA,
                             outline=outline,
                             attributes=attributes)


def derive_structure(personality, dropdown_act, outline):
    BASE_TEMPLATE_DERIVE_STRUCTURE = """Given this idea, refine this into story plot in the form of a {dropdown_act}.
    
    Idea:

    {outline},
 
    
    Come up with specific detail for each of the acts above. Add notes for each event that happens. 
    Pay special attention to what happens, where, how, and why."""

    return generate_response(personality,
                             BASE_TEMPLATE_DERIVE_STRUCTURE,
                             outline=outline,
                             dropdown_act=dropdown_act)


def refine_acts(personality, act, dropdown_refine2):
    attributes = ', '.join([str(elem) for i, elem in enumerate(dropdown_refine2)])
    BASE_TEMPLATE_REFINE_ACTS = """Refine this idea:
    '{act}'
    with the goal of making it more detailed and more {attributes}.
    Pay special attention to what happens, where, how, and why."""
    return generate_response(personality,
                             BASE_TEMPLATE_REFINE_ACTS,
                             act=act,
                             attributes=attributes)


def derive_chapters(personality, act, force_large_model=True):
    BASE_TEMPLATE_DERIVE_CHAPTERS = """Turn these story acts into chapters of a novel. 
    Add detailed notes on what happens, where, how, and why.
    You should generate at least 15-20 chapters. 
    Story:
    {act},
    Continue to generate until the whole list of chapters have been rewritten.
    """
   
        
    return generate_response(personality,
                             BASE_TEMPLATE_DERIVE_CHAPTERS,
                             act=act)
 
def command(playground):
    BASE_TEMPLATE_PLAYGROUND = """
    {playground}"""
    return generate_response("Generic", BASE_TEMPLATE_PLAYGROUND, playground=playground)
```
